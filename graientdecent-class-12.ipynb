```python
n = 39
itr = 50

for i in range(itr):
    predy = m*x + c
    Dm = (-2/n) * sum(x*(y-predy))
    Dc = (-2/n) * sum((y-predy))
    
    #update
    m = m - l*Dm
    c = c - l*Dc
    print(m,c)
```
